## Reward Functions

Priority is given to formatting of the reasoning and the correctness of the answers.

```python
PERFECT = 3
PARTIAL = 0.5
WRONG = -1
SIMILARITY_THRESHOLD = 0.8
NUMERIC_TOLERANCE = 0.2


def ensure_list(val, n:int) -> list:
    try:
        return list(val)
    except:
        return [None]*n

def normalize_text(text: str) -> str:
    """Normalize text for comparison."""
    if not text:
        return ""
    return " ".join(text.lower().strip().split())


def is_numeric(text: str) -> bool:
    """Check if text represents a number."""
    try:
        float(text.strip())
        return True
    except (ValueError, AttributeError):
        return False


def text_similarity(text1: str, text2: str) -> float:
    """Calculate similarity between two texts (0.0 to 1.0)."""
    if not text1 or not text2:
        return 0.0

    norm1 = normalize_text(text1)
    norm2 = normalize_text(text2)

    if norm1 == norm2:
        return 1.0

    if norm1 in norm2 or norm2 in norm1:
        return 0.9

    return SequenceMatcher(None, norm1, norm2).ratio()

match_format = re.compile(
    rf"^[\s]{{0,}}"
    rf"{REASONING_START}.+?{REASONING_END}.*?"
    rf"{SOLUTION_START}(.+?){SOLUTION_END}"
    rf"[\s]{{0,}}$",
    flags=re.MULTILINE | re.DOTALL,
)

answer_pattern = re.compile(
    rf"{SOLUTION_START}(.+?){SOLUTION_END}",
    flags=re.MULTILINE | re.DOTALL
)


def extract_answer(completion):
    """Extract answer from completion."""
    match = match_format.search(completion)
    if match:
        return match.group(1).strip()
    
    fallback_match = answer_pattern.search(completion)
    if fallback_match:
        return fallback_match.group(1).strip()
    
    return None


def match_format_exactly(prompts, completions, **kwargs):
    return [
        PERFECT if match_format.search(completion) else WRONG
        for completion in completions
    ]


def match_format_approximately(prompts, completions, **kwargs):
    scores = []
    for response in completions:
        score = 0
        is_whitespace = lambda s: not s or s.isspace()
        score += PARTIAL if response.count(REASONING_START) == 1 else WRONG
        score += PARTIAL if response.find(REASONING_START) == 0 else WRONG
        score += PARTIAL if response.count(REASONING_END) == 1 else WRONG

        if response.count(REASONING_END) == 1 and response.count(SOLUTION_START) == 1:
            r_end_idx = response.find(REASONING_END) + len(REASONING_END)
            s_start_idx = response.find(SOLUTION_START)
            if r_end_idx <= s_start_idx and is_whitespace(response[r_end_idx:s_start_idx]):
                score += PARTIAL
        else:
            score += WRONG

        score += PARTIAL if response.count(SOLUTION_START) == 1 else WRONG
        
        if response.count(SOLUTION_END) == 1:
            s_end_idx = response.find(SOLUTION_END) + len(SOLUTION_END)
            if is_whitespace(response[s_end_idx:]):
                score += PARTIAL * 2 # One for tag, one for no trailing text
            else:
                score += PARTIAL
        else:
            score += WRONG
            
        scores.append(score)
    
    return scores


def evaluate_numeric_answer(predicted, ground_truth):
    """Evaluate numeric answers with tolerance."""
    try:
        pred_num = float(predicted)
        true_num = float(ground_truth)
        
        if pred_num == true_num:
            return PERFECT
        
        if true_num == 0:
            return WRONG
        
        ratio = pred_num / true_num
        if 1 - NUMERIC_TOLERANCE <= ratio <= 1 + NUMERIC_TOLERANCE:
            return PARTIAL
        
        return WRONG
    except:
        return WRONG


def evaluate_text_answer(predicted, ground_truth):
    """Evaluate text answers with similarity."""
    similarity = text_similarity(predicted, ground_truth)
    
    if similarity >= 0.95:
        return PERFECT
    if similarity >= SIMILARITY_THRESHOLD:
        return PARTIAL
    
    return WRONG


def reward_gt_similarity_rapidfuzz(predicted, ground_truth):
    ratio_score = fuzz.ratio(predicted.lower(), ground_truth.lower()) / 100.0
    partial_ratio = fuzz.partial_ratio(predicted.lower(), ground_truth.lower()) / 100.0
    token_sort = fuzz.token_sort_ratio(predicted.lower(), ground_truth.lower()) / 100.0
    token_set = fuzz.token_set_ratio(predicted.lower(), ground_truth.lower()) / 100.0

    combined_score = (
        token_set * 0.4 +
        token_sort * 0.3 +
        partial_ratio * 0.2 + 
        ratio_score * 0.1
    )
    
    if combined_score >= 0.7:
        final_score = PERFECT
    elif combined_score >= 0.3:
        final_score = PARTIAL
    else:
        final_score = WRONG
        
    return final_score


def reward_readability_textstat(completion):
    score = 0.0
    
    if len(completion.strip()) < 50:
        return WRONG
    
    try:
        # Flesch Reading Ease: 0-100, higher = easier
        # Target: 60-80 (standard reading level)
        flesch_ease = textstat.flesch_reading_ease(completion)
        if 60 <= flesch_ease <= 80:
            score += PARTIAL
        
        # Flesch-Kincaid Grade Level: U.S. grade level
        # Target: 6-10 (readable but not simplistic)
        grade_level = textstat.flesch_kincaid_grade(completion)
        if 6 <= grade_level <= 10:
            score += PARTIAL
        
        # Lexicon count (total words)
        word_count = textstat.lexicon_count(completion, removepunct=True)
        if word_count >= 100:
            score += PARTIAL
        
        sentence_count = textstat.sentence_count(completion)
        if sentence_count >= 5:
            score += PARTIAL

        
    except Exception as e:
        return WRONG
        
    
    return score

def evaluate_creative_answer(predicted, ground_truth):
    return (reward_readability_textstat(predicted) + reward_gt_similarity_rapidfuzz(predicted, ground_truth))/2


def evaluate_math_science_answer(predicted, ground_truth):
    """Evaluate math/science answers (numeric or text)."""
    if is_numeric(predicted) and is_numeric(ground_truth):
        return evaluate_numeric_answer(predicted, ground_truth)
    
    return evaluate_text_answer(predicted, ground_truth)


def count_passing_tests(solution_code, test_code, entry_point, helpers):
    """Count proportion of tests that pass."""
    test_lines = [line.strip() for line in test_code.split('\n') if 'assert' in line]
    
    if not test_lines:
        return WRONG
    
    passed = 0
    for test_line in test_lines:
        try:
            namespace = {}
            exec(helpers, namespace)
            exec(solution_code, namespace)
            if '().' in entry_point:
                class_name, method_name = entry_point.split('().')
                obj = namespace[class_name]()
                entry_point_func = getattr(obj, method_name)
            else:
                entry_point_func = namespace[entry_point]
            test_statement = test_line.replace('candidate', 'func')
            namespace['func'] = entry_point_func
            exec(test_statement, namespace)
            passed += 1
        except:
            pass
    
    proportion = passed / len(test_lines)
    
    if proportion == 1.0:
        return PERFECT
    if proportion >= 0.5:
        return PARTIAL
    
    return WRONG


def evaluate_code_answer(solution_code, test_code, entry_point, helpers):
    """Evaluate code solution by running tests."""
    if not test_code or not entry_point or not helpers:
        return WRONG

    if len(solution_code.strip()) < 50:
        return WRONG
    
    try:
        namespace = {}
        exec(helpers, namespace)
        exec(solution_code, namespace)
        
        if '().' in entry_point:
            class_name, method_name = entry_point.split('().')
            obj = namespace[class_name]()
            entry_point_func = getattr(obj, method_name)
        else:
            entry_point_func = namespace[entry_point]
        
        exec(test_code, namespace)
        namespace['check'](entry_point_func)
        
        return PERFECT
        
    except AssertionError:
        return count_passing_tests(solution_code, test_code, entry_point, helpers)
        
    except Exception:
        return WRONG


def evaluate_answer_by_domain(predicted, ground_truth, domain, test_code, entry_point, helpers):
    if domain == 'code':
        return evaluate_code_answer(predicted, test_code, entry_point, helpers)
    if domain == "creative_writing":
        return evaluate_creative_answer(predicted, ground_truth)
    
    return evaluate_math_science_answer(predicted, ground_truth)

def check_answer_correct(prompts, completions, answer=None, domain=None, test_code=None, entry_point=None, helpers=None, **kwargs):
    scores = []
    n = len(completions)
    domains = domain
    test_codes = ensure_list(test_code, n)
    entry_points = ensure_list(entry_point, n)
    helpers_list = ensure_list(helpers, n)

    for idx, (completion, ground_truth) in enumerate(zip(completions, answer)):
        predicted = extract_answer(completion)        
        if predicted is None:
            scores.append(WRONG)
            continue
        
        score = evaluate_answer_by_domain(
            predicted=predicted,
            ground_truth=ground_truth,
            domain=domains[idx],
            test_code=test_codes[idx],
            entry_point=entry_points[idx],
            helpers=helpers_list[idx]
        )
        scores.append(score)

    return scores

REWARD_FUNCTIONS = [
    match_format_exactly,
    match_format_approximately,
    check_answer_correct,
]
```

```python
def convert_to_grain(hf_dataset, key_mapping, batch_size=DATAPROCESSING_BATCH_SIZE):
    data = []
    for item in hf_dataset:
        entry = {}
        for grain_key, hf_key in key_mapping.items():                         
              value = item.get(hf_key)                                          
              entry[grain_key] = value if value is not None else ''             
        data.append(entry)                                           

    return grain.MapDataset.source(data).batch(batch_size=batch_size)
```

```python
def generate(prompts, sampler, max_generation_steps=1, temperature=0.7, top_k=50, top_p=0.95, seed=None):
    if isinstance(prompts, str):
        input_batch = [prompts]
    else:
        input_batch = prompts

    out_data = sampler(
        input_strings=input_batch,
        max_generation_steps=max_generation_steps,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        echo=False,
        seed=seed if seed is not None else None,
        eos_tokens=EOS_TOKENS,
    )

    output = out_data.text
    if isinstance(prompts, str):
        return output[0]
    return output


def evaluate(
    dataset,
    sampler,
    temperature=0.7,
    top_k=50,
    top_p=0.95,
):
    reward_functions = REWARD_FUNCTIONS
    total_reward = 0.0
    total_count = 0
    max_steps = EVAL_MAX_STEPS

    for batch in tqdm(dataset):
        answers = batch["answer"]
        prompts = batch["prompts"]
        domains = batch.get("domain", [None]*len(prompts))
        test_codes = batch.get("test_code", [None]*len(prompts))
        entry_points = batch.get("entry_point", [None]*len(prompts))
        helpers_list = batch.get("helpers", [None]*len(prompts))

        responses = generate(
            prompts, sampler, max_steps, temperature, top_k, top_p
        )

        batch_rewards = []
        for reward_fn in reward_functions:
            rewards = reward_fn(
                prompts, 
                responses, 
                answer=answers,
                domain=domains,
                test_code=test_codes,
                entry_point=entry_points,
                helpers=helpers_list
            )
            batch_rewards.append(rewards)

        for i in range(len(responses)):
            total_reward_for_response = sum(
                batch_rewards[j][i] for j in range(len(reward_functions))
            )
            total_reward += total_reward_for_response
            total_count += 1
        if total_count % 10 == 0:
            avg_reward = total_reward / total_count
            print(
                f"===> Evaluated {total_count} examples | "
                f"Avg Reward: {avg_reward:.2f}"
            )
    avg_reward = total_reward / total_count if total_count > 0 else 0.0

    return avg_reward


TEST_QUERIES = [

    # Math questions
    "If a store sells apples for $2 each and oranges for $3 each, how much would 5 apples and 3 oranges cost in total?",
    "A rectangle has a length of 12 cm and a width of 8 cm. What is its area?",
    "John has 20 candies. He gives 1/4 of them to his friend. How many candies does John have left?",
    
    # Science questions
    "What is the chemical symbol for water?",
    "Is photosynthesis a chemical change or a physical change?",
    "What type of energy does a battery store?",

    # Creative writing questions
    "Write a short poem about the changing seasons.",
    "Describe a futuristic city in 100 words.",
    "Create a dialogue between two characters who have just met for the first time.",

    # Coding questions
    "Write a Python function that checks if a number is prime.",
    "How would you implement a binary search algorithm in Java?",
    "Explain the concept of recursion with an example in C++."
]


def test_model_with_queries(model, queries=TEST_QUERIES):
    """Test model with sample queries across different domains."""
    test_sampler = sampler_lib.Sampler(
        transformer=model,
        tokenizer=tokenizer,
        cache_config=sampler_lib.CacheConfig(
            cache_size=EVAL_CACHE_SIZE,
            num_layers=model_config.num_layers,
            num_kv_heads=model_config.num_kv_heads,
            head_dim=model_config.head_dim,
        ),
    )

    for i, query in enumerate(queries, 1):
        print(f"\nQuery {i}/{len(queries)}:")
        print(f"Q: {query}")

        # Format the query with system prompt
        formatted_query = TEMPLATE.format(
            system_prompt=SYSTEM_PROMPT,
            question=query
        )

        response = generate(
            formatted_query,
            test_sampler,
            max_generation_steps=EVAL_MAX_STEPS,
            temperature=0.7,
            top_k=50,
            top_p=0.95
        )

        print(f"A: {response}")

    print(f"\n{'_/!\\'*20}\n")


def evaluate_model(model, dataset):
    """Evaluate model on dataset using reward functions."""

    grain_mapping = {
        'prompts': 'prompt',
        'question': 'question',
        'answer': 'answer',
        'domain': 'domain',
        'test_code': 'test_code',
        'entry_point': 'entry_point',
        'helpers': 'helpers',  
    }

    grain_dataset = convert_to_grain(dataset, grain_mapping, batch_size=EVAL_BATCH_SIZE)

    sampler = sampler_lib.Sampler(
        transformer=model,
        tokenizer=tokenizer,
        cache_config=sampler_lib.CacheConfig(
            cache_size=EVAL_CACHE_SIZE,
            num_layers=model_config.num_layers,
            num_kv_heads=model_config.num_kv_heads,
            head_dim=model_config.head_dim,
        ),
    )

    avg_reward = evaluate(
        grain_dataset,
        sampler,
        temperature=1e-4,
        top_k=1,
        top_p=1.0
    )

    return avg_reward
```

```python


def get_adamw_optimizer(warmup_steps, decay_steps, max_learning_rate):
    optimizer = optax.adamw(
        learning_rate=optax.schedules.warmup_cosine_decay_schedule(
            init_value=0.0,
            peak_value=max_learning_rate,
            warmup_steps=warmup_steps,
            decay_steps=decay_steps,
            end_value=0.0,
        ),
        b1=ADAMW_B1,
        b2=ADAMW_B2,
        weight_decay=ADAMW_WEIGHT_DECAY,
    )

    optimizer = optax.chain(
        optax.clip_by_global_norm(max_norm=ADAMW_MAX_GRAD_NORM),
        optimizer,
    )

    return optimizer


def train_grpo(model, dataset, checkpoint_dir):
    """Train model using GRPO and return updated model."""

    grain_mapping = {
        'prompts': 'prompt',
        'question': 'question',
        'answer': 'answer',
        'domain': 'domain',
        'test_code': 'test_code',
        'entry_point': 'entry_point',
        'helpers': 'helpers',
    }

    grain_dataset = convert_to_grain(dataset, grain_mapping)

    grpo_optimizer = get_adamw_optimizer(
        warmup_steps=GRPO_WARMUP_STEPS,
        decay_steps=GRPO_MAX_STEPS,
        max_learning_rate=GRPO_LEARNING_RATE,
    )

    grpo_checkpointing_options = ocp.CheckpointManagerOptions(
        save_interval_steps=SAVE_INTERVAL_STEPS, max_to_keep=MAX_TO_KEEP
    )

    grpo_metrics_logging_options = metrics_logger.MetricsLoggerOptions(
        log_dir=f"{BASE_CKPT_DIR}/tmp/tensorboard/grpo", flush_every_n_steps=20
    )

    grpo_cluster_config = rl_cluster_lib.ClusterConfig(
        role_to_mesh={
            rl_cluster_lib.Role.ACTOR: mesh,
            rl_cluster_lib.Role.REFERENCE: mesh,
            rl_cluster_lib.Role.ROLLOUT: mesh,
        },
        rollout_engine='vanilla',
        offload_to_cpu=GRPO_OFFLOAD_REF_TO_CPU,
        training_config=rl_cluster_lib.RLTrainingConfig(
            actor_optimizer=grpo_optimizer,
            eval_every_n_steps=EVAL_EVERY_N_STEPS,
            max_steps=GRPO_MAX_STEPS,
            mini_batch_size=GRPO_BATCH_SIZE,
            train_micro_batch_size=GRPO_BATCH_SIZE,
            metrics_logging_options=grpo_metrics_logging_options,
            checkpoint_root_directory=checkpoint_dir,
            checkpointing_options=grpo_checkpointing_options,
        ),
        rollout_config=base_rollout.RolloutConfig(
            max_tokens_to_generate=GRPO_MAX_OUTPUT_LENGTH,
            max_prompt_length=GRPO_MAX_PROMPT_LENGTH,
            kv_cache_size=GRPO_MAX_CACHE_SIZE,
            temperature=GRPO_TEMPERATURE,
            top_p=GRPO_TOP_P,
            top_k=GRPO_TOP_K,
            eos_tokens=EOS_TOKENS,
        ),
    )

    grpo_config = GRPOConfig(
        num_generations=GRPO_NUM_GENERATIONS,
        num_iterations=GRPO_NUM_ITERATIONS,
        beta=GRPO_BETA,
        epsilon=GRPO_EPSILON,
    )

    grpo_cluster = rl_cluster_lib.RLCluster(
        actor=model,
        reference=ref_model,
        tokenizer=tokenizer,
        cluster_config=grpo_cluster_config,
    )

    grpo_learner = GRPOLearner(
        rl_cluster=grpo_cluster,
        reward_fns=REWARD_FUNCTIONS,
        algo_config=grpo_config,
    )

    with mesh:
        grpo_learner.train(grain_dataset, None)
        
    return model
```
